```python
import pandas as pd
#loading the dataset
dataset=pd.read_csv("heart_disease_prediction.csv")
```

```python
#splitting target and features
x=dataset.drop(columns='Heart Disease',axis=1)
y=dataset['Heart Disease']
```

```python
#splitting into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

*The data is split into 80% and 20% as train data and test data.*

```python
#training the model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(solver='lbfgs',max_iter=1000)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
#evaluation of performance
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
x_train_prediction = model.predict(x_train.values)
training_data_accuracy = accuracy_score(x_train_prediction,y_train.values)
print('accuracy of training data:',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test.values)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test.values)
print('accuracy of testing data:',testing_data_accuracy)
```

```python
confusion=confusion_matrix(y_test.values,x_test_prediction)
print('confusion matrix is:',confusion)
```

```python
report=classification_report(y_test.values,x_test_prediction)
print('classifiction report is:',report)
```

* accuracy of the model is 80%,meaning it correctly predicted the presence or absence of heart disease in 80% of the cases.
* The precision for predicting absence is higher (0.81) than for predicting presence (0.78), indicating that the model is slightly more reliable in predicting patients without heart disease.
* The recall for predicting presence is lower (0.75) than for absence (0.83), suggesting that there were more false negatives in predicting heart disease.
* The F1-score, which balances precision and recall, is also higher for absence predictions.
